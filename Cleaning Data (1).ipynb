```python
import numpy as np
import pandas as pd 
```

```python
# WE START THE ANALYSIS WITH THE YEARLY DATASET
# Read the CSV file
file_path = 'housing_in_london_yearly_variables.csv'
df = pd.read_csv(file_path)

# Replace 'nan' with empty string
df.replace('nan', '', inplace=True)

# Write cleaned data to a new CSV file
cleaned_file_path = 'housing_in_london_yearly_variables_cleaned.csv'
df.to_csv(cleaned_file_path, index=False)

print("Data cleaning complete. Cleaned data saved to:", cleaned_file_path)
```
